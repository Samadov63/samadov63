```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\csv\StudentsPerformance.csv')
```

```python
df.head()
```

```python
df['final_score'] = ((df['math score'] + df['reading score'] + df['writing score'])/3).round(1)
```

```python
df
```

```python
df['grade 5'] = df['final_score'] > 80 
```

```python
df
```

```python
df.head()
```

```python
total_grade = []
for i in df['final_score']:
    if i >= 80:
        total_grade.append(5)
    if (i < 80) and (i >=65):
        total_grade.append(4)
    if (i < 65) and (i >= 50):
        total_grade.append(3)
    if(i < 50) and (i >= 1):
        total_grade.append(2)
    
df['total_grade'] = total_grade 
```

```python
df.drop(columns = ['grade 5', 'grade 4', 'grade 3', 'grade 2'], axis = 1, inplace = True )
```
