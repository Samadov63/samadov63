```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt



df=pd.read_csv(r'C:\Users\HP\Desktop\FIFA 2018 Statistics.csv')
```

```python
df.fillna({'1st Goal': 0}, inplace = True)
df.fillna({'Own goals': 0}, inplace = True)
df.fillna({'Own goal Time': 0}, inplace = True)
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df, columns = ['Team', 'Opponent', 'Team', 'Round', 'PSO', 'Man of the Match', 'Date'])
```

```python
y=df1['Goal Scored']
X=df1.drop('Goal Scored',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
rfc=RandomForestClassifier().fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
rfc.score(X_test,y_test)
```
